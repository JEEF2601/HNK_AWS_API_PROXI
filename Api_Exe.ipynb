```python
import importlib
import json


# Importa el m√≥dulo lambda_function desde la carpeta api
lambda_module = importlib.import_module("lambda_function")

class LambdaContext:
    def _init_(self):
        self.function_name = "test_function"
        self.memory_limit_in_mb = 128
        self.invoked_function_arn = "arn:aws:Test"
        self.aws_request_id = "1234-5678-9101"

def invoke_lambda(event):

    context = LambdaContext()
    response = lambda_module.lambda_handler(event, context)
    print("Respuesta Lambda:", response)
    return response

_name_ = "_main_"
if _name_ == "_main_":
    # Define un evento de prueba
    test_event = {
        "key1": "valor1",
        "key2": "valor2",
        "key3": "valor3"
    }
```
